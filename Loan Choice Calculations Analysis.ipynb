# Calculating optimal loans

What really matters:
    - Inputs:
        1) Income
        2) House Value
        3) Outstanding Costs
        4) Outstanding Debt
        5) Loan term
        6) Down Payment
    - Outputs
        1) How much interest will be payed
        2) How much will monthly payments cost
        


- Values of interest
    - Gross Debt Service Ratio (GDS)
        - Monthly housing-related costs and divide it by your gross monthly income
        - bellow 39 %
    - Total Debt Service Ratio (TDS)
        - Monthly debt related costs and divide it by your gross monthly income
        - bellow 44 %
    - Loan To Value Ration (LTV)
        - the size of the loan to the value of the property.
